## Test enrichment in each UTR

```python
pwms = pd.read_hdf('../../VirusHostInteractionAtlas/DATA/PWMs/attract_rbpdb_encode_filtered_human_pwms.h5')
```

```python
model = smeagol.models.PWMModel(pwms)
```

```python
genome = smeagol.io.read_fasta('../../VirusHostInteractionAtlas/DATA/Genomes/Genomes/V/Monopartite/Filoviridae/GCF_000848505.1/GCF_000848505.1_ViralProj14703_genomic.fna.gz')
```

```python
enrs = {}
for name, c in coords.items():
    print(name)
    seq = [genome[0][c[0]:c[1]]]
    enr = smeagol.enrich.enrich_in_genome(seq, model, simN=1000, simK=2, 
                                                    rcomp='none', threshold=0.8, 
                                                    combine_seqs=True, background='binomial')['enrichment']
    enrs[name] = enr[enr.fdr<.05]
```
